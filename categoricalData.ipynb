Write a Python program to convert categorical data (e.g., Yes/No) into numerical values (1/0) and display the updated dataset.

```python
import numpy as np
import pandas as pd 
```

```python
df = pd.DataFrame(columns=['StudentID', 'Study_Hours', 'Marks', 'Good'])

df['StudentID'] = ['S1','S2','S3','S4','S5']
df['Study_Hours'] = [2, 3, None, 5, 6]
df['Marks'] = [35, 40, 45, None, 70]
df['Good'] = ["Yes", "No", "Yes", "Yes", "No"]

catMap = {
    'Yes' : 1,
    'No' : 0
}
df['Good'] = df['Good'].map(catMap)

df
```
